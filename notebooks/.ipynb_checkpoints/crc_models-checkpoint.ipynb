## Make a sample dataset

```python
# Make sample dataset:
def make_sample(data_loc, out_loc, downsample_factor):
    data_classes = glob.glob(data_loc+'/*')
    for data_class_loc in data_classes:
        imgs = glob.glob(data_class_loc+'/*')
        data_class = data_class_loc.rsplit('/', 1)[1]
        num = len(imgs)
        sample_num = int(num/downsample_factor)
        samp = np.random.choice(imgs, sample_num)
        for file in samp:
            name = file.rsplit('/', 1)[1]
            new_loc = os.path.join(out_loc, data_class)
            if not os.path.exists(new_loc):
                os.makedirs(new_loc)
            copyfile(file, os.path.join(new_loc , name))

# data_loc = '/home/rbbidart/project/rbbidart/cancer_hist/crc_128/valid'
# out_loc = '/home/rbbidart/project/rbbidart/cancer_hist/incp_128_sample_valid'
# make_sample(data_loc, out_loc, 50)
```

## Train a few epochs to see if it is working
1. Check with no augmentation
2. Check with augmentation

```python
sys.path.insert(0, '/home/rbbidart/cancer_hist/src')
from heat_models import*

import keras
from keras import backend as K
from keras.engine.topology import Layer
from keras.layers import Dropout, Flatten, Reshape, Input
from keras.layers.core import Activation, Dense, Lambda
from keras.callbacks import ModelCheckpoint, EarlyStopping
from keras.preprocessing.image import ImageDataGenerator


im_size = 128
out_loc = '/home/rbbidart/cancer_hist_out/crc/incp_128_sample'
train_loc = '/home/rbbidart/project/rbbidart/cancer_hist/incp_128_sample'
valid_loc = '/home/rbbidart/project/rbbidart/cancer_hist/incp_128_sample_valid'
num_train = len(glob.glob(train_loc+'/**/*', recursive=True))
num_valid = len(glob.glob(valid_loc+'/**/*', recursive=True))
print('num_train', num_train)
print('num_valid', num_valid)


parameters = {
'learning_rate': .0002, 
'dropout': .1,
'im_size': int(im_size),
'num_out': 6 
}

# Params for all models
epochs=100
batch_size=32   # make this divisible by len(x_data)
steps_per_epoch = np.floor(num_train/batch_size) # num of batches from generator at each epoch. (make it full train set)
validation_steps = np.floor(num_valid/batch_size) # size of validation dataset divided by batch size
print('validation_steps', validation_steps)
image_shape = (int(im_size), int(im_size))

def triangular2(epoch):
    base_lr = .0005
    max_lr = .005
    step_size = 5
    cycle = np.floor(1+epoch/(2*step_size))
    x = np.abs(epoch/step_size - 2*cycle + 1)
    lr = base_lr + (max_lr-base_lr)*np.maximum(0, (1-x))/float(2**(cycle-1))
    return lr

train_datagen = ImageDataGenerator(
    rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        train_loc,
        target_size=image_shape,
        batch_size=batch_size,
        class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(
        valid_loc,
        target_size=image_shape,
        batch_size=batch_size,
        class_mode='categorical')

model = conv_incp3(**parameters)
name = 'conv_incp3'+'_'+str(im_size)
out_file=os.path.join(str(out_loc), name)

callbacks = [EarlyStopping(monitor='val_loss', patience=15, verbose=0),
    keras.callbacks.LearningRateScheduler(triangular2),
    ModelCheckpoint(filepath=os.path.join(out_loc, name + '_.{epoch:02d}-{val_acc:.2f}.hdf5'), 
        verbose=1, monitor='val_loss', save_best_only=True)]

hist = model.fit_generator(train_generator,
                              validation_data=validation_generator,
                              steps_per_epoch=int(steps_per_epoch), 
                              epochs=epochs,
                              validation_steps=int(validation_steps),
                              callbacks=callbacks)
pickle.dump(hist.history, open(out_file, 'wb'))

data=hist.history
print(name, 'Final 5 Epochs Validation loss: ', np.mean(data['val_acc'][-5:]))
print(name, 'Epoch 100 avg Validation loss: ', np.mean(data['val_acc'][100:105]))
plt.plot(data['loss'])    
plt.plot(data['acc'])
plt.plot(data['val_acc'])
plt.title(name)
plt.ylabel('Accuracy')
plt.xlabel('epoch')
plt.legend(['Train loss', 'Train acc', 'val_acc'], loc='upper left')
plt.figure(figsize=(40,10))
plt.show()
```
