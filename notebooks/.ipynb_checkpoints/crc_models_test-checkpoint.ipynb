### Adjusting functions to work on CRC data:

```python
def create_heatmap_crc(image_loc, model, height, downsample):
    def softmax(x):
        e_x = np.exp(x - np.max(x))
        return e_x / e_x.sum()
    image = np.asarray(Image.open(image_loc))
    image_shape = image.shape
    image = image/255.0 # During training the images were normalized
    height = int(height)

    last = model.layers[-2].output
    model = Model(model.input, last)

    out_shape = np.ceil(np.array(image.shape)/float(downsample)).astype(int)
    out_shape[2] = 6

    delta=int((height)/2)
    image = np.lib.pad(image, ((delta, delta-int(downsample)), (delta, delta-int(downsample)), (0,0)), 'constant', constant_values=(0, 0))
    image = np.expand_dims(image, axis=0)
    heat = model.predict(image, batch_size=1, verbose=0)
    heat = np.reshape(heat, out_shape)
    heat[:,:,0:] = np.apply_along_axis(softmax, 2, heat[:,:,0:])
    return heat

def non_max_supression_up(heatmap, radius=5, cutoff=.5, stride=2, output_class=True): 
    # initial ordering of heatmaps is dir ordering: [detection, epithelial, fibriblast, inflammatory, negative, others]
    # Make the negative samples first, and detection second (don't care about detection for classification)
    heatmap[:, :, [0, 1, 2, 3, 4, 5]] = heatmap[:, :, [4, 0, 1, 2, 3, 5]]
    # final ordering: [negative, detection, epithelial, fibriblast, inflammatory, others]
    
    # Upsample the image to the original shape before doing non-max suppression
    labels = []
    points = []    
    heatmap[:, :, 0] = 1-heatmap[:, :, 0]
    heatmap = scipy.ndimage.zoom(heatmap, int(stride), order=1)
    heatmap = np.lib.pad(heatmap, ((radius, radius), (radius, radius), (0,0)), 'constant', constant_values=(0, 0))

    curr_max = 1
    while (curr_max > float(cutoff)):
        max_coord = np.asarray(np.unravel_index(heatmap[:, :, 0].argmax(), heatmap[:, :, 0].shape))
        # Add the maximum cell probability to the coordinate
        if output_class: 
            max_coord = np.append(max_coord, np.argmax(heatmap[max_coord[0], max_coord[1], 2:])) 
        elif not output_class:
            max_coord = np.append(max_coord, heatmap[max_coord[0], max_coord[1], :])
            
        # find the max set all classes within radius r to p = 0
        curr_max = heatmap[:, :, 0].max()
        for row in range(-1*radius, radius, 1):
            for col in range(-1*radius, radius, 1):
                # dont't just do a square
                dist = np.sqrt(row** 2 + col** 2)
                if (dist<=radius):
                    heatmap[int(max_coord[0]+row), int(max_coord[1]+col), :] = 0
        # adjust for the padding that was added
        max_coord[0] = max_coord[0] - radius
        max_coord[1] = max_coord[1] - radius
        
        points.append(max_coord)
    points = np.array(points)
    #points = points[points[:,2]!=0]
    
    points = points.astype(float)
    return points

def get_matched_pts2_crc(point_dict, preds, cor_rad=6, mode='detection'):
    all_matched_pts = []
    all_matched_preds = []
    total_nuclei = 0
    num_predicted = 0
    true_points = []

    #loop through the predictions, and check if there a corresponding true point
    # Delete the true points once they are matched, so they are not mached more than once
    if(mode == 'detection'):
        true_points = np.array(point_dict['detection'])
    else:
        # use numbering:
        # detection 9
        # epithelial 0
        # fibroblast 1
        # inflammatory 2
        # others 3
        try:
            all_nuclei = point_dict['detection'].tolist()
        except:
            all_nuclei = np.vstack([point_dict['epithelial'], point_dict['fibroblast'], 
                                       point_dict['inflammatory'], point_dict['others']]).tolist()
            print('No detected point list')

        all_labeled_nuclei = np.vstack([point_dict['epithelial'], point_dict['fibroblast'], 
                                       point_dict['inflammatory'], point_dict['others']]).tolist()
        only_detection_nuclei = [row for row in all_nuclei if row not in all_labeled_nuclei]
        point_dict['9'] = only_detection_nuclei
        point_dict.pop('detection')
        point_dict['0'] = point_dict.pop('epithelial')
        point_dict['1'] = point_dict.pop('fibroblast')
        point_dict['2'] = point_dict.pop('inflammatory')
        point_dict['3'] = point_dict.pop('others')

        for nuclei_type, point_list in point_dict.items():
            for point in point_list:
                true_points.append([point[0], point[1], nuclei_type])
        
        if (len(true_points) !=len(all_nuclei)):
            print('len(true_points)', len(true_points))
            print('len(all_nuclei)', len(all_nuclei))
            
    true_points = np.array(true_points).astype(float)
    true_points_pred = np.c_[true_points, np.repeat(99, true_points.shape[0])]  
    tp_temp = true_points

    for index, point in enumerate(preds):
        dists = np.sqrt(np.sum((point[0:2] - tp_temp[:, 0:2]) ** 2, axis=1))
        if (len(dists)==0):
            break
        else:
            min_ind = np.argmin(dists)
            # if point has matching prediction, append it and increment the number of matched points
            if (dists[min_ind] < cor_rad):
                all_matched_preds.append(preds[index, 2:]) 
                all_matched_pts.append(tp_temp[min_ind, :])
                # Add to all points
                if ( mode!='detection'):
                    true_ind = np.where((true_points_pred[:, :2] == tp_temp[min_ind, :2]).all(axis=1))
                    true_points_pred[true_ind[0], 3] = tp_temp[min_ind, 2]
                    if(len(true_ind)>1):
                        print('multiple points matched')
                tp_temp = np.delete(tp_temp, (min_ind), axis=0) # If the point is matched, delete from list
    acc_dict = {}
    acc_dict["all_matched_preds"] = np.array(all_matched_preds)
    acc_dict["all_matched_pts"] = np.array(all_matched_pts)
    acc_dict["num_matched"] = len(all_matched_pts)
    acc_dict["total_nuclei"] = len(true_points)
    acc_dict["num_predicted"] = len(preds)
    acc_dict["abs_error"] = abs(len(true_points)-len(preds))
    acc_dict["true_points_pred"] = true_points_pred
    return acc_dict   
```

```python
model_loc = '/home/rbbidart/cancer_hist_out/crc/incp3_64_f1/conv_incp3_64_.23-0.72.hdf5'

parameters = {
'learning_rate': .0002, 
'dropout': .1,
'im_size': 64,
'num_out': 6 
}
model = conv_incp3(**parameters)
model.load_weights(model_loc)

valid_loc = '/home/rbbidart/project/rbbidart/cancer_hist/crc_ttv_split_f1/valid/images'
all_imgs=glob.glob(valid_loc+'/**/*', recursive=True)

label_locs = '/home/rbbidart/project/rbbidart/cancer_hist/crc_ttv_split_f1/valid/labels'
all_labels=glob.glob(os.path.join(label_locs, '*'))
all_labels = [loc for loc in all_labels if loc.rsplit('.', 1)[0][-1] == 'm']

cor_rad = 6
cutoff_list = [.7, .8, .85, .9, .95, .99]
radius_list = [8, 10, 12]


for radius in radius_list:
    for cutoff in cutoff_list:
        all_matched_pts = []
        all_matched_preds = []
        abs_error_list = []
        total_nuclei = 0
        num_predicted = 0
        num_matched = 0

        for image_loc in all_imgs:
            img_name = image_loc.rsplit('/', 1)[1].rsplit('.', 1)[0]
            if img_name in ['img86']:
                continue
            # get the labels. Find matching label image, and extract the coordinates:
            label_loc = image_loc.replace('images', 'labels')
            label_loc = label_loc.replace('.bmp', '.pickle')
            point_dict = pickle.load(open(label_loc, "rb" ))
            image = np.asarray(Image.open(image_loc))

            heat = create_heatmap_crc(image_loc, model=model, height=64, downsample=2)
            heat[:, :, [0, 1, 2, 3, 4, 5]] = heat[:, :, [4, 1, 2, 3, 0, 5]]

            preds = non_max_supression_up(heat, radius=radius, cutoff=cutoff, stride=2)
            preds[:,[0, 1]] = preds[:,[1, 0]]
            
            acc = get_matched_pts2_crc(point_dict, preds, cor_rad=cor_rad)
            
            # Update 
            all_matched_preds.extend(np.array(acc["all_matched_preds"]))
            all_matched_pts.extend(np.array(acc["all_matched_pts"]))
            num_matched = num_matched + acc['num_matched']
            total_nuclei = total_nuclei + acc['total_nuclei']
            num_predicted = num_predicted + acc['num_predicted']
            abs_error_list.append(acc['abs_error'])

        precision = num_matched/num_predicted # (detected&true/detected)
        recall = num_matched/total_nuclei # (detected&true/true)
        f1 = 2*(precision*recall)/(precision+recall)
        print('------------------------')
        print('radius: ', radius, 'cutoff: ', cutoff)
        print('total_nuclei: ', total_nuclei, 'num_matched: ', num_matched,  'num_predicted: ', num_predicted)
        print('% identified: ', num_matched/total_nuclei)
        print('% of preds that are false positive: ', (num_predicted-num_matched)/num_predicted)
        print('precision', precision)
        print('recall', recall)
        print('f1', f1)
```

```python
def test_detection(model, img_size, test_loc, label_locs, radius, cutoff, cor_rad=6):
    all_imgs=glob.glob(test_loc+'/**/*', recursive=True)
    all_labels=glob.glob(os.path.join(label_locs, '*'))
    all_labels = [loc for loc in all_labels if loc.rsplit('.', 1)[0][-1] == 'm']

    all_matched_pts = []
    all_matched_preds = []
    abs_error_list = []
    total_nuclei = 0
    num_predicted = 0
    num_matched = 0


    for image_loc in all_imgs:
        img_name = image_loc.rsplit('/', 1)[1].rsplit('.', 1)[0]
        if img_name in ['img86']:
            continue
        # get the labels. Find matching label image, and extract the coordinates:
        label_loc = image_loc.replace('images', 'labels')
        label_loc = label_loc.replace('.bmp', '.pickle')
        point_dict = pickle.load( open(label_loc, "rb" ))
        image = np.asarray(Image.open(image_loc))

        heat = create_heatmap_crc(image_loc, model=model, height=img_size, downsample=2)
        # switch the dimensions so that the nuclei probability is first

        preds = non_max_supression_up(heat, radius=radius, cutoff=cutoff, stride=2)
        # preds is ordered in the opposite way to the points from point_dict (x, y)
        preds[:,[0, 1]] = preds[:,[1, 0]]

        acc = get_matched_pts2_crc(point_dict, preds, cor_rad=cor_rad, mode='detection')

        # Update 
        all_matched_preds.extend(np.array(acc["all_matched_preds"]))
        all_matched_pts.extend(np.array(acc["all_matched_pts"]))
        num_matched = num_matched + acc['num_matched']
        total_nuclei = total_nuclei + acc['total_nuclei']
        num_predicted = num_predicted + acc['num_predicted']
        abs_error_list.append(acc['abs_error'])

    precision = num_matched/num_predicted # (detected&true/detected)
    recall = num_matched/total_nuclei # (detected&true/true)
    f1 = 2*(precision*recall)/(precision+recall)

    print('------------------------')
    print('radius: ', radius, 'cutoff: ', cutoff)
    print('total_nuclei: ', total_nuclei, 'num_matched: ', num_matched,  'num_predicted: ', num_predicted)
    print('% identified: ', num_matched/total_nuclei)
    print('% of preds that are false positive: ', (num_predicted-num_matched)/num_predicted)
    print('precision', precision)
    print('recall', recall)
    print('f1', f1)
```

# Classification Accuracy

```python
def test_class_2stage(model_1, model_2, img_size, test_loc, label_locs, radius, cutoff, cor_rad=6):
    # Use the 64 model like normal to give the output, then classify with 128
    # No efficiency here
    
    all_imgs=glob.glob(test_loc+'/**/*', recursive=True)
    all_labels=glob.glob(os.path.join(label_locs, '*'))
    all_labels = [loc for loc in all_labels if loc.rsplit('.', 1)[0][-1] == 'm']

    all_matched_pts = []
    all_matched_preds = []
    abs_error_list = []
    true_points_pred = []
    all_preds = []
    total_nuclei = 0
    num_predicted = 0
    num_matched = 0

    for image_loc in all_imgs[0:5]:
        img_name = image_loc.rsplit('/', 1)[1].rsplit('.', 1)[0]
        if img_name in ['img86']:
            continue
        # get the labels. Find matching label image, and extract the coordinates:
        label_loc = image_loc.replace('images', 'labels')
        label_loc = label_loc.replace('.bmp', '.pickle')
        point_dict = pickle.load( open(label_loc, "rb" ))
        image = np.asarray(Image.open(image_loc))

        heat = create_heatmap_crc(image_loc, model=model_1, height=img_size, downsample=2)
        # switch the dimensions so that the nuclei probability is first

        preds = non_max_supression_up(heat, radius=radius, cutoff=cutoff, stride=2)
        # preds is ordered in the opposite way to the points from point_dict (x, y)
#         preds[:,[0, 1]] = preds[:,[1, 0]]
        # Update all predictions using the 128 model
        preds = predict_points(preds, model_2, image, im_size=128, heat=False)
        preds[:,[0, 1]] = preds[:,[1, 0]]
        acc = get_matched_pts2_crc(point_dict, preds, cor_rad=cor_rad, mode='class')

        # Update 
        all_matched_preds.extend(np.array(acc["all_matched_preds"]))
        all_matched_pts.extend(np.array(acc["all_matched_pts"]))
        num_matched = num_matched + acc['num_matched']
        total_nuclei = total_nuclei + acc['total_nuclei']
        num_predicted = num_predicted + acc['num_predicted']
        abs_error_list.append(acc['abs_error'])
        true_points_pred.extend(acc['true_points_pred'])
        all_preds.extend(preds)

    precision = num_matched/num_predicted # (detected&true/detected)
    recall = num_matched/total_nuclei # (detected&true/true)
    f1 = 2*(precision*recall)/(precision+recall)

    print('------------------------')
    print('radius: ', radius, 'cutoff: ', cutoff)
    print('total_nuclei: ', total_nuclei, 'num_matched: ', num_matched,  'num_predicted: ', num_predicted)
    print('% identified: ', num_matched/total_nuclei)
    print('% of preds that are false positive: ', (num_predicted-num_matched)/num_predicted)
    print('precision', precision)
    print('recall', recall)
    print('f1', f1)
    
    acc = {}
    acc["all_matched_preds"] = np.array(all_matched_preds)
    acc["all_matched_pts"] = np.array(all_matched_pts)
    acc["num_matched"] = np.array(num_matched)
    acc["total_nuclei"] = total_nuclei
    acc["num_predicted"] = num_predicted
    acc["abs_error"] = abs_error_list
    acc["true_points_pred"] = np.array(true_points_pred)
    acc["all_preds"] = np.array(all_preds)
    return acc
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import itertools

print(accuracy_score(acc["all_matched_pts"][:,2], acc["all_matched_preds"]))
print(accuracy_score(acc['true_points_pred'][:,2], acc['true_points_pred'][:,3]+1))

cm = confusion_matrix(acc["all_matched_pts"][:,2], acc["all_matched_preds"])
classes =['epithelial', 'fibroblast', 'inflammatory', 'others']

plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
# add the numbers
thresh = cm.max() / 2
for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j, i, cm[i, j],
             horizontalalignment="center",
             color="white" if cm[i, j] > thresh else "black")
plt.colorbar()
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45)
plt.yticks(tick_marks, classes)
plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()


# weighted f1 score for classification
weights = {'epithelial':7722/22444, 'fibroblast':5712/22444, 'inflammatory':6971/22444, 'others':2039/22444}

matched = {'epithelial':np.count_nonzero(acc["true_points_pred"][:, 3] == 0), 
           'fibroblast':np.count_nonzero(acc["true_points_pred"][:, 3] == 1), 
           'inflammatory':np.count_nonzero(acc["true_points_pred"][:, 3] == 2), 
           'others':np.count_nonzero(acc["true_points_pred"][:, 3] == 3)}

total = {'epithelial':np.count_nonzero(acc["true_points_pred"][:, 2] == 1), 
         'fibroblast':np.count_nonzero(acc["true_points_pred"][:, 2] == 2), 
          'inflammatory':np.count_nonzero(acc["true_points_pred"][:, 2] ==3), 
          'others':np.count_nonzero(acc["true_points_pred"][:, 2] == 4)}

predicted = {'epithelial':np.count_nonzero(acc["all_preds"][:, 2] == 1), 
           'fibroblast':np.count_nonzero(acc["all_preds"][:, 2] == 2), 
           'inflammatory':np.count_nonzero(acc["all_preds"][:, 2] == 3), 
           'others':np.count_nonzero(acc["all_preds"][:, 2] == 4)}

precision = {}
recall = {}
f1 = {}

for nuclei_type, count in matched.items():
    precision[nuclei_type] = matched[nuclei_type]/predicted[nuclei_type] # (detected&true/detected)
    recall[nuclei_type] = matched[nuclei_type]/total[nuclei_type]# (detected&true/true)
    f1[nuclei_type] = 2*(precision[nuclei_type]*recall[nuclei_type])/(precision[nuclei_type]+recall[nuclei_type])
    
print(precision)
print(recall)
print(f1)
print(weights)


weighted_f1 = 0
for nuclei_type, count in f1.items():
    weighted_f1 = weighted_f1+weights[nuclei_type]*f1[nuclei_type]
print('weighted_f1', weighted_f1)
```
